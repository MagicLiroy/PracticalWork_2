```python
class Chocolate:
  def __init__(self, price, additive):
    if price <= 0:
      raise ValueError('Цена шоколадки меньше или равна нулю!')
    else:
      self.price = price

    self.additive = additive

    if additive == None:
      self.additive = None
    else:
      self.additive = additive

  def show_my_chocolate(self):
    if self.additive != '':
      print(f'Шоколадный батончик, добавка: {self.additive}')
    else:
      print('Обычный шоколадный батончик')

  def __str__(self):
    return f'Шоколадный батончик, {self.price} рублей'

chocolate1 = Chocolate(10.12, 'ваниль')
chocolate1.show_my_chocolate()
print(chocolate1)

chocolate2 = Chocolate(10.12, '')
chocolate2.show_my_chocolate()
print(chocolate2)

chocolate3 = Chocolate(0, '')
chocolate3.show_my_chocolate()
print(chocolate3)


```
